```python
import csv
import os
import pandas

data = pandas.read_csv('velocity-data-1.csv', header=-1)
data
```

```python
import matplotlib.pyplot as plt
import numpy as np

pos = data[0].as_matrix()
V = data[2].as_matrix()

vel = np.zeros_like(pos)
for i in np.arange(1, pos.shape[0]-1):
    vel[i] = (pos[i+1]-pos[i-1])/0.02
    
# vel = vel[1:-1]
display(vel)

vel = vel[1:-1]/1000
V = V[1:-1]

# x = vel[1:-1]/10
# y = V[1:-1]

plt.plot(vel, V, '.')
plt.show()
```

```python
import scipy.stats
x = vel
y = V
result = scipy.stats.linregress(x, y)
display(result)
print(dir(result))
```

```python
x = vel
y = V

i = np.where(np.abs(x)<5)[0]

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.scatter(x[i], y[i], alpha=0.8, s=1)

_x = np.array([np.min(x[i]), np.max(x[i])])
_y = result.slope*_x+result.intercept
ax.plot(_x, _y)
print('R-value: {:.3f}'.format(result.rvalue))
plt.show()
```

```python
rvalues = []
for i in range(50):
    x = vel[i:]
    if i == 0:
        y = V
    else:
        y = V[:-i]
    result = scipy.stats.linregress(x, y)
    rvalues.append((i, result.rvalue))
    
rvalues = np.array(rvalues)
j = np.argmax(rvalues[:,1])


x = rvalues[:,0]
y = rvalues[:,1]
plt.plot(x, y)
plt.plot(x[j], y[j], 'o')
plt.grid()
```

r value is maximized when the data is shifted by 12 points

Need to still consider if the data should be shifted in the other direction

```python
rvalues = []
for i in range(50):
    y = V[i:]
    if i == 0:
        x = vel
    else:
        x = vel[:-i]
    result = scipy.stats.linregress(x, y)
    rvalues.append((i, result.rvalue))
    
rvalues = np.array(rvalues)
j = np.argmax(rvalues[:,1])


x = rvalues[:,0]
y = rvalues[:,1]
plt.plot(-x, y)
plt.plot(-x[j], y[j], 'o')
plt.grid()
```

Ok so the fit is definitely maximized for shift 12

```python
x = vel[12:]
y = V[:-12]

result = scipy.stats.linregress(x, y)
display(result)

# i = np.where(np.abs(x)<5)[0]

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.scatter(x, y, alpha=0.8, s=1)

_x = np.array([np.min(x), np.max(x)])
_y = result.slope*_x+result.intercept
ax.plot(_x, _y)
ax.set_xlabel('Velocity (m/s)')
ax.set_ylabel('V')
print('R-value: {:.3f}'.format(result.rvalue))
plt.show()
```
